## De-duplicating

```python
fn_dedup = os.path.join(sc.path_data,f'matches.{sc.name}--{sc.name}.csv')
if not os.path.exists(fn_dedup):
    sc.match_records(sc,year_window=None)
    mdf_dedup = pd.read_csv(fn_dedup).sort_values('match_ratio',ascending=False).fillna('')
    mdf_dedup = mdf_dedup[mdf_dedup.id<mdf_dedup.id2].drop_duplicates(['id'],keep='first')
    mdf_dedup.to_csv(fn_dedup,index=False)
else:
    mdf_dedup=pd.read_csv(fn_dedup)
mdf_dedup
```

### Manual intervention

```python
fn_dedup_anno=fn_dedup.replace('.csv','.anno.csv')
if not os.path.exists(fn_dedup_anno): 
    mdf_dedup.to_csv(fn_dedup_anno,index=False)
```

```python
# re-load
mdf_dedup = pd.read_csv(fn_dedup_anno).fillna('')
mdf_dedup
```

```python
df_dups=mdf_dedup[mdf_dedup.match_ismatch=='y']
df_dups
```

## Matching

### With "LitLab" corpus

```python
match_fn = os.path.join(sc.path_data,'matches.SemanticCohort--LitLab.csv')
if not os.path.exists(match_fn):
    sc.match_records(ll,year_window=None)
    mdf=pd.read_csv(match_fn)
    mdf=mdf[mdf.match_ratio_author>=70]
    mdf=mdf.sort_values('match_ratio',ascending=False).fillna('').drop_duplicates(['id'],keep='first')
    mdf.to_csv(match_fn,index=False)
else:
    mdf=pd.read_csv(match_fn)
mdf
```

### Manual intervention

```python
match_fn_anno=match_fn.replace('.csv','.anno.csv')
if not os.path.exists(match_fn_anno):
    mdf.to_csv(match_fn_anno,index=False)
```

```python
mdf=pd.read_csv(match_fn_anno).query('match_ismatch == "y"')
mdf
```

```python
def shorten_title(
        title,
        prfxs=['St.','Mr.','Mrs.','Dr.','Rev.'],
        splits=[':',';',', or',', a ',', A ','/','.','[','(']):
    for x in prfxs:
        title=title.replace(x+' ', x[:-1]+' ')
    for x in splits:
        title=title.split(x)[0].strip()
    return title
```
